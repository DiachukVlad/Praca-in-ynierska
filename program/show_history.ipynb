```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def show(history):
    print(max(history['val_accuracy']))
    f = plt.figure()
    f.set_figwidth(10)
    f.set_figheight(3.5)
    plt.plot(history['accuracy'])
    plt.plot(history['val_accuracy'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'val'], loc='upper left')
    plt.xticks(range(len(history)))
    plt.grid(color='gray', linestyle='-', linewidth=.5)
    plt.show()
        
    f = plt.figure()
    f.set_figwidth(10)
    f.set_figheight(3)
    plt.plot(history['loss'])
    plt.plot(history['val_loss'])
    plt.title('model loss')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'val'], loc='upper left')
    plt.xticks(range(len(history)))
    plt.grid(color='gray', linestyle='-', linewidth=.5)
    plt.show()
```

```python
df = pd.read_csv('history/history6.csv')
df = df.drop(df.index[15])

df1 = pd.read_csv('history/history6_1.csv')
df1 = df1.drop(df1.index[range(7,9)])

df2 = pd.read_csv('history/history6_2.csv')

full_data = pd.concat([df, df1, df2], ignore_index=True)

show(full_data)
```
